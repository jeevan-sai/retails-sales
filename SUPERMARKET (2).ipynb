```python
cols=[ 'Tax 5%','Total','cogs', 'gross income']
```

```python
for i in cols:
    skew=df[i].skew().round()
    print(skew,"-->",i)
```

```python
for i in cols:
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iqr=q3-q1
    upper=q3+1.5*iqr
    lower=q1-1.5*iqr
    df[i]=df[i].map(lambda x:  upper if x>upper  else lower if x<lower else x)
```

```python
sns.boxplot(df)
plt.xticks(rotation=45)
plt.show()
```
