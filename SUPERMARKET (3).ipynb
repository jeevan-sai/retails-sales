```python
cols=[ 'Tax 5%','Total','cogs', 'gross income']
```

```python
for i in cols:
    skew=df[i].skew().round()
    print(skew,"-->",i)
```

```python
for i in cols:
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iqr=q3-q1
    upper=q3+1.5*iqr
    lower=q1-1.5*iqr
    df[i]=df[i].map(lambda x:  upper if x>upper  else lower if x<lower else x)
```

```python
sns.boxplot(df)
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
L=LabelEncoder()
L.fit(df[["Gender"]])
```

```python
L.transform(df[["Gender"]])
```

```python
df["Gender"]=L.transform(df[["Gender"]])
```

```python
df["Payment"]=df["Payment"].replace({'Ewallet':0,'Cash':1,'Credit card':2})
```

```python
df.drop("Invoice ID",axis=1,inplace=True)
```

```python
df
```

```python
df["Date"]=pd.to_datetime(df["Date"])
```

```python
df["Time"]=pd.to_datetime(df["Time"],format='%H:%M').dt.time
```

```python
df["Day of week"]=df["Date"].dt.day_name()
```

```python
df["month"]=df["Date"].dt.month
```

```python
df["Hour"]=pd.to_datetime(df["Time"],format="%H:%M:%S").dt.hour
```

```python
df["profit margin"]=(df["Unit price"]*df["Quantity"]-df["Tax 5%"])
```

### EDA

```python
df.groupby("Product line")["Total"].sum().plot(kind="bar",color="r",width=0.2)
plt.xticks(rotation=60)
plt.title("productline")
plt.show()
```

```python
##customer segement--analysing customer type
avg_cust_type=df.groupby(df["Customer type"])["Total"].mean()
avg_cust_type.plot(kind="bar",color="skyblue")
plt.title("average spend by customer type")
plt.xlabel("customer type")
plt.ylabel("avg_spend")
plt.show()
```

```python
sales_trend=df.groupby('Date')["Total"].sum()
sales_trend.plot()
plt.title("sales trend over time")
plt.xlabel("date")
plt.ylabel("total sales")
plt.show()
```

```python
##payment analysis
payment_count=df["Payment"].value_counts().plot(kind="bar")
plt.title("payment count")
plt.ylabel("count")
plt.xlabel("payment method")
plt.show()
```

### performance analysis

```python
sale_branch=df.groupby('Branch')["Total"].sum().plot(kind="bar",color="g")
plt.title("total sales by branch")
plt.xlabel("branch")
plt.ylabel("total sales")
plt.show()
```
